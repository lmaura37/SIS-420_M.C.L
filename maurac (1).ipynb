```python
def nextStates(current):
	nodes=[]

	for action in POSSIBILITIES:
		
		nextState = deepcopy(current)
		nextState.prev=current
		
		#El barco estará en el lado opuesto.
		nextState.boat = 1-current.boat
		
		#De izquierda a derecha
		if(current.boat==0):

			#Aumentar el número en el lado derecho
			nextState.right[0]+=action[0]
			nextState.right[1]+=action[1]
			
			#Disminuye el número en el lado izquierdo
			nextState.left[0]-=action[0]
			nextState.left[1]-=action[1]
		
		#De derecha a izquierda
		elif(current.boat==1):
			
			#Disminuye el número en el lado derecho
			nextState.right[0]-=action[0]
			nextState.right[1]-=action[1]
			
			#Aumentar el número en el lado izquierdo
			nextState.left[0]+=action[0]
			nextState.left[1]+=action[1]
		
		if nextState.isValidStateCheck():
			nodes.append(nextState)

	return nodes
```

```python
def bfs(roots):
	
	if roots.GoalState():
		return roots
	
	visited = set()
	queue = [roots]

	while queue:
		state = queue.pop()
		if state.GoalState():
			return state
		
		visited.add(state)
		
		for child in nextStates(state):
			if child in visited:
				continue

			if child not in queue:
				queue.append(child)
```

```python
from copy import deepcopy

POSSIBILITIES = [[1,1],[0,2],[2,0],[0,1],[1,0]] # Posibilidades 

class State():
	
	def __init__(self, left, boat, right): #Base Class`
		self.left=left;
		self.boat = boat;
		self.right=right;
		self.prev = None
		
	def __str__(self):
		return("({},{}) - ({},{}) - {}".format(self.left[0],self.left[1], self.right[0],self.right[1], self.boat))
	
	#Comprobar restricciones
	def isValidStateCheck(self):	
		#Para ambos bancos, los misioneros presentes en el banco no pueden ser superados en número por caníbales.
		if(0 < self.left[0] < self.left[1] or 0 < self.right[0] < self.right[1]):
			return False	
		
		#No se transportan más misioneros/caníbales de los que existen en un lado
		if(self.left[0]<0 or self.left[1]<0 or self.right[0]<0 or self.right[1]<0):
			return False
		
		return True

	def __eq__(self, other):
		return (self.left[0]==other.left[0] and self.left[1] == other.left[1] and self.right[0]==other.right[0] and self.right[1]==other.right[1] and self.boat==other.boat)
		
	def __hash__(self):
		return hash((self.left[0],self.left[1],self.boat,self.right[0],self.right[1]))
	
	def __str__(self):
		return("({},{}) - ({},{}) - {}".format(self.left[0],self.left[1],self.right[0],self.right[1],self.boat))
	
	#Mueve con éxito a todos los misioneros y caníbales del lado izquierdo al derecho
	def GoalState(self):
		return(self.left[0]==0 and self.left[1]==0)
```
